```python
import pandas as pd
import seaborn as smb
import numpy as np

# allow graphs to display in notebook iteself
%matplotlib inline
```

```python
# give the paths to all data files available

data_file = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-08 to 2018-05-15_integrated.csv", index_col=[0])
data_file_1 = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-16 to 2018-05-23_integrated.csv", index_col=[0])
data_file_2 = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-24 to 2018-05-31_integrated.csv", index_col=[0])

data_file = pd.concat([data_file, data_file_1, data_file_2])
```

```python
data_file
```

```python
specific_link = data_file.loc[data_file.ID ==2340]
specific_link.reset_index()
```

```python
maximum_series = []
```

```python
maximum = -10000000000000000000000000000000000000
for index in specific_link["RSL_MAX"]:
    if index >= maximum:
        maximum = index
    maximum_series.append(maximum)
specific_link["MAX_OF_RSL_MAX"] = maximum_series
```

```python
specific_link["MAX_OF_RSL_MAX"]
```

```python
n = np.arange(1, len(specific_link) +1)
```

```python
specific_link["samples"] = n
```

```python
specific_link["MAX_OF_RSL_MAX_MINUS_TSL"] = specific_link["TSL_MAX"].sub(specific_link["MAX_OF_RSL_MAX"])
smb.pairplot(specific_link[:96], x_vars=["samples"], y_vars=["MAX_OF_RSL_MAX"], size=5, aspect=2)
```
