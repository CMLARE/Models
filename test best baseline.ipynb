```python
import pandas as pd
from matplotlib import pyplot as plot
import seaborn as sb

# allow plots to appear in the notebook
%matplotlib inline
```

```python
data_file = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-24 to 2018-05-31.csv", index_col=["ID","DateTime"], parse_dates=["DateTime"])
```

```python
data_file_new = data_file.loc[892]
```

```python
data_file_new
```

```python
#     array to store all the baseline values
baseline = []

#     since the CML values are consistent(No missing values in the middle) we can do as following
length_of_df = len(data_file_new)

#     calculate baseline for first 24 items
# Here we take the next 6 hours
for i in range(0, 11):
    baseline_val = data_file_new[i:i+24]["RSL_MAX"].max()
    baseline.append(baseline_val)

#    calculate baseline for middle elements
#  Here we take the forward 3 hours and backward 3 hours(in middle)
for i in range(11, length_of_df - 11):
    baseline_val = data_file_new[i-11:i+13]["RSL_MAX"].max()
    baseline.append(baseline_val)

#    calculate baseline for last 12 elements     
#  Here we consider the last 6 hours
for i in range(length_of_df-11, length_of_df):
    baseline_val = data_file_new[i-24:i]["RSL_MAX"].max()
    baseline.append(baseline_val)

data_file_new["Baseline"] = baseline
```

```python
data_file_new["Baseline_ATTN"] = data_file_new["Baseline"].sub(data_file_new["RSL_MIN"])
```
