Creating a simple model to predict the rainfall. this is for testing purposes. this model only considered the rainy periods and predicts the results

# Data

RainFall data : Department of Meteorology - Sri Lanka.

Microwave link data: Dialog Axiata, Mobile network operator company [Sri Lanka]

THIS MODEL USES RSL_MIN VALUES AND FREQUNCY, PATH LENGTH AS THE FEATURES. THERE IS NO BASELINE CORRECTION FOR THIS MODEL

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import cross_val_score
```

```python
data_file = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-08 to 2018-05-15_integrated.csv")
data_file_1 = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-16 to 2018-05-23_integrated.csv")
data_file_2 = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-24 to 2018-05-31_integrated.csv")

data_file = pd.concat([data_file, data_file_1, data_file_2])
```

```python
data_file
```

```python
data_file_less_5 = data_file.loc[lambda x: x["distance"] < 5]
```

Select RSL_MIN, FREQUENCY, PATH LENGTH as Features in the model

```python
features = ["RSL_MIN", "PathLength"]
```

```python
x_data = data_file_less_5[features]
```

```python
y_data = data_file_less_5["precipitation(mm)"]
```

```python
y_data.shape
x_data.shape
```

```python
random_forest_regressor = RandomForestRegressor(n_estimators=500)
```

dataset is divided into 1:10 as test_data:training_data
Used the k-fold cross validaion to check wheter the model is accurate

```python
length_fixed_to_10_multiple = int(len(x_data)/10) * 10
length_fixed_to_10_multiple
```

```python
x_data = x_data[:length_fixed_to_10_multiple]
y_data = y_data[:length_fixed_to_10_multiple]
```

```python
scores = cross_val_score(random_forest_regressor, x_data,y_data, cv=10, scoring="neg_mean_absolute_error")
print(scores)
```

```python
scores.mean()
```
