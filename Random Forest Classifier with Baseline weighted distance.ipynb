```python
data_file_less_5 = new_data.loc[lambda x: x["distance"] < 5]

data_file_less_5["Baseline_ATTN"] = data_file_less_5["Baseline"].sub(data_file_less_5["RSL_MIN"])

features = ["refined", "PathLength"]

def calculate_time(row):
    if (row["RSL_MAX"] - row["RSL_MIN"]) == 0:
        return 15
    else:
        time  = 15*(row["RSL_MAX"] - row["RSL_AVG"]) / (row["RSL_MAX"] - row["RSL_MIN"])
        return time
    
data_file_less_5["refined"] = data_file_less_5.apply(lambda x: calculate_time(x)*x["Baseline_ATTN"], axis=1)
data_file_less_5["time"] = data_file_less_5.apply(lambda x: calculate_time(x), axis=1)

x_data_new = data_file_less_5[features]
y_data_new = data_file_less_5["class"]

y_predicted = random_forest_classifier.predict(x_data_new)
```
