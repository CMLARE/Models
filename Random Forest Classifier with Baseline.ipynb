```python
import pandas as pd
import pickle as pk
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
```

```python
data_file = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-08 to 2018-05-15_integrated_classified.csv")
data_file_1 = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-16 to 2018-05-23_integrated_classified.csv")
data_file_2 = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-05-24 to 2018-05-31_integrated_classified.csv")
data_file_3 = pd.read_csv("/home/cmlare/Data/RF Data/Processed/2018-06-01 to 2018-06-10_integrated_classified.csv")

data_file = pd.concat([data_file, data_file_1, data_file_2, data_file_3])
```

Select RSL_MIN, FREQUENCY, PATH LENGTH as Features in the model

```python
# data_file["hour"] = pd.to_datetime(data_file["date_time"]).dt.hour
data_file["Baseline_ATTN"] = data_file["Baseline"].sub(data_file["RSL_MIN"])
data_file["BaselineA_PathLen"] = data_file.apply(lambda x: x["Baseline_ATTN"]/x["PathLength"], axis=1)
```

```python
data_file
```

```python
def calculate_time(row):
    if (row["RSL_MAX"] - row["RSL_MIN"]) == 0:
        return 15
    else:
        time  = 15*(row["RSL_MAX"] - row["RSL_AVG"]) / (row["RSL_MAX"] - row["RSL_MIN"])
        return time
```

```python
data_file["refined"] = data_file.apply(lambda x: calculate_time(x)*x["Baseline_ATTN"], axis=1)
data_file["time"] = data_file.apply(lambda x: calculate_time(x), axis=1)
```

```python
features = ["Baseline_ATTN","PathLength","FrequencyBand", "RSL_MAX", "TSL_MAX", "RSL_AVG"]
```

```python
x_data = data_file[features]
```
